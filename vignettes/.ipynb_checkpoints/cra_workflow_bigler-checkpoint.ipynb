# Data Download

Since the dataset is huge, we can not provide it on Github but you can download it (huge data set!)

```python
outpath = Path("data-raw")
url = ("https://ncar-cesm2-lens.s3-us-west-2.amazonaws.com/catalogs/aws-cesm2-le.json")
catalog = intake.open_esm_datastore(url)
```

```python
catalog.df
```

```python
var = ['Q', 'T', 'PSL']
scenarios = ['historical']
catalog_subset = catalog.search(variable = var, experiment = scenarios, frequency = 'daily', forcing_variant = 'cmip6')
```

```python
catalog_subset.df
```

```python
try:
    # Use to_dataset_dict to load the data from the filtered catalog subset
    datasets = catalog_subset.to_dataset_dict(storage_options={'anon': True})
    print("Data loaded successfully!")

    # Optionally, save data to disk
    for var_name, var_data in datasets.items():
        # Saving the data to NetCDF files
        var_data.to_netcdf(f"{var_name}_data.nc")
        print(f"Saved {var_name} data to NetCDF.")
except Exception as e:
    print("An error occurred while loading the dataset:", e)
```
